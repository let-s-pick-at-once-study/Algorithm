    def __init__(self):
        self.container = []
        self.count = 0
        
    def push(self, value):
        self.container.append(value)
        self.count += 1
    
    def pop(self):
        if self.empty():
            return -1
        self.count -= 1
        return self.container.pop()
    
    def size(self):
        return self.count
    
    def empty(self):
        if self.count == 0:
            return 1
        return 0
    
    def top(self):
        if self.empty():
            return -1
        return self.container[-1]
